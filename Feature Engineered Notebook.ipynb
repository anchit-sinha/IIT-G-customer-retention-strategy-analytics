# **Missing Value Imputaion**

```python
df.isnull().sum()
```

```python
df['RAF'] = np.where(df['Review Rating'].isna(),0,1)
```

```python
category_median = (df.groupby("Category")["Review Rating"].transform("median"))
```

```python
df["Review Rating"] = (df["Review Rating"].fillna(category_median))
```

# **Discount Binary Creation**

```python
df['Disc'] = np.where(df['Discount Applied'] == 'Yes',1,0)
```

```python
df["NoDisc"] = 1 - df['Disc']
```

# **Subscription Binary**

```python
df['Sub'] = np.where(df['Subscription Status'] == 'Yes',1,0)
```

# **Normalization of Base Metrics**

**Frequency Score**

```python
freq_map = {'Weekly':7, 'Fortnightly':6, 'Bi-Weekly':5, 'Monthly':4, 'Quarterly':3, 'Every 3 Months':2, 'Annually':1}
```

```python
df['Freq_Score'] = (df['Frequency of Purchases'].map(freq_map))
```

**Age Group Scoring**

```python
def age_group(age):
    if age <= 25:
        return 'Young'
    elif age <= 35:
        return 'Emerging'
    elif age <= 50:
        return 'Established'
    else:
        return 'Mature'
```

```python
df['Age Group'] = df['Age'].apply(age_group)
```

**Tenure Tier**

```python
def tenure_tier(pp):
    if pp <= 12:
        return 'New'
    elif pp <= 25:
        return 'Growing'
    elif pp <= 37:
        return 'Established'
    else:
        return 'Veteran'
```

```python
df['Tenure_Tier'] = df['Previous Purchases'].apply(tenure_tier)
```

**PA_norm**

```python
df['PA_norm'] = (df['Purchase Amount (USD)'] - df['Purchase Amount (USD)'].min()) / (df['Purchase Amount (USD)'].max() - df['Purchase Amount (USD)'].min())
```

**PP_norm**

```python
df['PP_norm'] = (df['Previous Purchases'] - df['Previous Purchases'].min()) / (df['Previous Purchases'].max() - df['Previous Purchases'].min())
```

**Freq_norm**

```python
df['Freq_norm'] = (df['Freq_Score'] - df['Freq_Score'].min()) / (df['Freq_Score'].max() - df['Freq_Score'].min())
```

**Rating_norm**

```python
df['Rating_norm'] = (df['Review Rating'] - df['Review Rating'].min()) / (df['Review Rating'].max() - df['Review Rating'].min())
```

# **CORE METRICS**

### *VALUE METRICS*

**Customer Value Score (CVS)**

```python
df['CVS'] = (0.5 * df['PA_norm'] + 0.3 * df['PP_norm'] + 0.2 * df['Freq_norm'])
```

**Historical Revenue Proxy (HRP)**

```python
df['HRP'] = (df['Purchase Amount (USD)'] * df['Previous Purchases'])
```

### *LOYALTY METRICS*

**Behavioral Loyalty Score A (BLS-A)**

```python
df['BLS_A'] = (0.5 * df['PP_norm'] + 0.3 * df['Freq_norm'] + 0.2 * df['Sub'])
```

**Behavioral Loyalty Score B (BLS-B)**

```python
df['BLS_B'] = (0.4 * df['PP_norm'] + 0.4 * df['Freq_norm'] + 0.2 * df['Sub'])
```

**Behavioral Loyalty Score C (BLS-C)**

```python
df['BLS_C'] = (0.3 * df['PP_norm'] + 0.5 * df['Freq_norm'] + 0.2 * df['Sub'])
```

**Quality Loyalty Score (QLS)**

```python
df['QLS'] = (.4 * df['Rating_norm'] + 0.3 * df['Sub'] + 0.3 * df['NoDisc'])
```

```python
df.head()
```

**Selecting one of the 3 BLS as the temporary BLS for now. Once SQL proves which one is the better, we will replace BLS_Temp = BLS_B with A or C**
*BLS_C came out to be the dominant one, hence we have taken BLS_TEMP = BLS_C now*

```python
df["BLS_TEMP"] = df["BLS_C"]
```

**FULL PRICE LOYALTY SCORE (FPLS)**

```python
df['FPLS'] = (0.5*df['BLS_TEMP'] + 0.5* df['NoDisc'])
```

# **RETENTION AND HEALTH METRICS**

**Customer Health Score (CHS)**

```python
df['CHS'] = (0.3 * df['Rating_norm'] + 0.4 * df['Freq_norm'] + 0.3 * df['PP_norm'])
```

**Satisfaction Score (SS)**

```python
df['SS'] = df['Rating_norm']
```

**SATISFACTION RISK INDEX (SRI)**

```python
df['SRI'] = 1 - df['Rating_norm']
```

**RETENTION POTENTIAL SCORE (RPS)**

```python
df['RPS'] = (0.5*df['CHS'] + 0.5*df['QLS'])
```

# **PROMOTION METRICS**

**PROMOTION USAGE FLAG (PUF)**

```python
df['PUF'] = df['Disc']
```

**PROMOTION INDEPENDENCE SCORE (PIS)**

```python
df['PIS'] = df['NoDisc']
```

**MARGIN PROTECTION SCORE (MPS)**

```python
df['MPS'] = (0.6*df['FPLS'] + 0.4*df['CVS'])
```

**PROMO SUNSET READINESS SCORE (PSRS)**

```python
df['PSRS'] = (0.4*df['PIS']+ 0.3*df['FPLS']+ 0.3*df['QLS'])
```

**CUSTOMER MATURITY SCORE (CMS)**

```python
df['CMS'] = (0.7*df['PP_norm'] + 0.3*df['Freq_norm'])
```

**FUTURE CHAMPION SCORE (FCS)**

```python
df['FCS'] = (0.5*df['CHS'] + 0.3*df['QLS'] + 0.2*df['CVS'])
```

**PREMIUM CUSTOMER SCORE (PCS)**

```python
df['PCS'] = (0.5*df['CVS'] + 0.3*df['FPLS'] + 0.2*df['CHS'])
```

**FEEDBACK ENGAGEMENT SCORE (FES)**

```python
df['FES'] = (0.7*df['RAF'] + 0.3*df['Sub'])
```

# **CUSTOMER SEGMENTATION**

**Segment Function**

```python
def customer_segment(row):
    if row['PCS'] >= 0.75:
        return 'Premium Customer'
    elif row['FCS'] >= 0.65:
        return 'Future Champion'
    elif row['PSRS'] >= 0.7:
        return 'Promo Sunset Candidate'
    elif row['SRI'] >= 0.6:
        return 'At Risk'
    else:
        return 'Standard Customer'
```

```python
df['Customer_Segment'] = df.apply(customer_segment, axis=1)
```
